```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def func(x, a, b, c, d):
    return a * b**(c * x + d)

times = np.array([4.14, 10.25, 27.92, 1377.814])
qubits = np.array([4, 6, 8, 12])


popt, pcov = curve_fit(func, qubits, times)


a, b, c, d = popt


x_values = np.linspace(min(qubits), 20, 20)  
y_values = func(x_values, a, b, c, d)


plt.scatter(qubits, times, label='Original Data')


plt.plot(x_values, y_values, 'r-', label=f'Fitted Curve: {a:.2f} * {b:.2f}^({c:.2f} * x + {d:.2f})')

plt.xlabel('Qubits')
plt.ylabel('Times')
plt.title('Fitted Curve for the Data')
plt.legend()
plt.grid(True)
plt.show()
''
```
